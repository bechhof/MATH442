## Research Project

Suppose we have \$1,000,000 for investment. 

We are considering four alternative mutual funds: A, B, C and D.
Based on historical data, analysis shows that the annual return rates for fund A, B, C and D are Unif$[-40\%, 60\%]$, Normal$(8\%, (6\%)^2)$, Normal$(7\%, (3\%)^2)$, and Unif$[6\%, 10\%]$, respectively.

The following are some possible questions which you may want to answer. 

(a) What is the expected return if you invest option A for 5 years? What is the probability that your profit is more than 25%? What is the probability that you lose 25% of your money?

(b) What are your answers if you invest fund B or C? And what is your recommendation for the
best option?

(c) How will your answer/recommendation be changed if you are planning for 10, 20, or 30 years, instead of 5 years?

(d) Will it be a better alternative if we invest \$500,000 for fund A and \$500,000 for fund D? Or how about \$600,000 for fund B and \$400,000 for fund C?

(e) Will it be a better alternative if we invest \$500,000 for fund A, \$300,000 for fund B and \$200,000 for fund C?



```python
%matplotlib inline
import numpy as np
from numpy.random import seed
from scipy.stats import norm, uniform
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
initial_investment = 1000000
```

```python
seed(102)

yearly_returns_A = uniform.rvs(-0.4, 0.6, size = (2000,5))
yearly_returns_A += 1

yearly_returns_compounded_A = yearly_returns_A.prod(axis=1)
outcomes_for_A = yearly_returns_compounded_A * initial_investment
```

```python
outcomes_for_A_plot = sns.distplot(outcomes_for_A, kde=False)
outcomes_for_A_plot.locator_params(axis='x',nbins=5)
plt.title("Outcomes for A after 5 years")
plt.savefig("Outcomes for A after 5 years.pdf")
plt.show()
```

```python
print("For A, the probability that the profit is more than 25%\
 is approximately {0:.4f}.".format(
        (yearly_returns_compounded_A > 1.25).mean()))
```

```python
print("For A, the probability that the loss is more than 25%\
 is approximately {0:.4f}.".format(
        (yearly_returns_compounded_A < 0.75).mean()))
```

```python
print("For A, the mean of the returns\
 is approximately {0:.4f}.".format(
        (-1+(yearly_returns_compounded_A).mean())))
```
